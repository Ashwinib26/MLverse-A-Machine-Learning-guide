### Logistic Regression Demonstration (Heart Dataset)

```python
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/cw/Downloads/Heart.csv")
```

```python
data.head(20)
```

```python
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
eChestPain = LabelEncoder()
eThal = LabelEncoder()
eAHD = LabelEncoder()
```

```python
data['num_ChestPain']=eChestPain.fit_transform(data['ChestPain'])
```

```python
data['num_AHD']=eAHD.fit_transform(data['AHD'])
```

```python
data['num_Thal']=eThal.fit_transform(data['Thal'])
```

```python
data.head(11)
```

```python
data.drop(columns=['Thal'],inplace=True)
```

```python
data.drop(columns=['ChestPain'],inplace=True)
```

```python
data.drop(columns=['AHD'],inplace=True)
```

```python
data.head(20)
```

```python
data.isnull().sum()
```

```python
datas = data.dropna()
```

```python
datas.isnull().sum()
```

```python
x=datas.drop(columns=['num_AHD'])
x
```

```python
y=datas['num_AHD']
```

```python
y
```

```python
y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaler = scaler.fit_transform(x_train)
```

```python
x_test_scaler = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
regr = LogisticRegression()
regr.fit(x_train_scaler,y_train)
```

```python
y_train_scaler_predict = regr.predict(x_train_scaler)
y_train_scaler_predict
```

```python
regr.score(x_train_scaler,y_train)
```

```python
y_test_scaler_predict = regr.predict(x_test_scaler)
y_test_scaler_predict
```

```python
regr.score(x_test_scaler,y_test)
```

```python
log_regr = LogisticRegression(random_state=23,C=1,fit_intercept=True)
#fit_intercept - intercept point considered or not , c -> biasing
log_regr.fit(x_train_scaler,y_train)
```

```python
log_regr.score(x_train_scaler,y_train)
```

```python
log_regr.score(x_test_scaler,y_test)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
conf_mat = confusion_matrix(y_test , y_test_scaler_predict)
```

```python
conf_mat
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(conf_mat,annot=True,fmt="d",cmap="Blues",cbar=False,xticklabels=['No Dissease','Disease'],yticklabels=['No Dissease','Disease'])
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.title('Confusion Matrix')

plt.show()
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
import matplotlib.pyplot
```

```python
print("Classification Report :")
print(classification_report(y_test,y_test_scaler_predict,target_names=['No Disease','Disease']))
```
